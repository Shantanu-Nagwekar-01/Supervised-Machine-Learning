```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
data = {
    'Hours': [1, 2, 3, 4.5, 5.5, 6, 7, 8.5, 9, 10],
    'Scores': [20, 22, 35, 45, 52, 60, 66, 78, 85, 95]
}

df = pd.DataFrame(data)
df
```

```python
plt.scatter(df['Hours'], df['Scores'], color = "indianred", marker='x')
plt.title("Study Hours vs Exam Score")
plt.xlabel("Hours Studied")
plt.ylabel("Score")
plt.grid(True)
plt.show()
```

```python
X = np.array(df["Hours"]).reshape(-1,1) #features 2d array
y = df["Scores"] #target 1d array
```

```python
print("Shape of X (features):", (X.shape)) 
print("Shape of y (targets):", (y.shape)) 
```

```python
model = LinearRegression()
model.fit(X, y)
```

```python
hours = 7.5
predicted_score = model.predict([[hours]])
print(f"Predicted Score for studying {hours} hours : {predicted_score[0]:.2f}")
```

```python
#Draw Regression Line
regression_line = model.coef_ * X + model.intercept_
plt.scatter(X, y, color = "darkblue", marker = 'x', label = "Actual Data")
plt.plot(X, regression_line, color="indianred", label = "Regression Line")
plt.title("Study Hours vs Exam Score")
plt.xlabel("Hours Studied")
plt.ylabel("Score")
plt.grid(True)
plt.legend()
plt.show()
```

```python
from sklearn.metrics import mean_squared_error,  r2_score
y_pred = model.predict(X)
print("Mean Squared Error: ", mean_squared_error(y, y_pred))
print("R2_Score: ", r2_score(y, y_pred))
```

```python
regression_line = model.coef_ * X + model.intercept_
plt.scatter(X, y, color = "darkblue", marker = 'x', label = "Actual Data")
plt.scatter(X, y_pred, color = "orange", marker = 'o', label = "Predicted_Values")
plt.plot(X, regression_line, color="teal", label = "Regression Line")
plt.title("Study Hours vs Exam Score")
plt.xlabel("Hours Studied")
plt.ylabel("Score")
plt.grid(True)
plt.legend()
plt.show()
```
